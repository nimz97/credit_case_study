## Finding top 10 correlations 

```python
#heat map of correlation for non-defaulters.
plt.figure(figsize=(12,8)) 
sns.heatmap(df_0.corr(), annot=True, cmap="YlGn")
plt.title('Correlation matrix for non-defaulters(target variable 0)')
plt.show()
```

```python
#finding top 10 correlations
corr0 = df_0.corr()
corr_df0 = corr0.where(np.triu(np.ones(corr0.shape), k=1).astype(np.bool))
corr_df0 = corr_df0.unstack().reset_index().dropna(subset = [0])
corr_df0.columns = ['Var1', 'Var2', 'Corr_Value']
corr_df0.sort_values(by = "Corr_Value", ascending =False, inplace = True)
corr_df0.head(10)
```

## Inferences from target 0

*  AMT_GOODS_PRICE and AMT_CREDIT shows a highly correlated , which implies that if the price of the goods for which the loan is given is higher then the Credit amount of their loan is also higher.
* AMT_ANNUITY and AMT_INCOME_TOTAL shows a high correlation, which implies that clients with high loan annuity has high annual income.
* AMT_ANNUITY and AMT_CREDIT are third highly correlated, that implies for high loan annuity, Credit amount of the loan is also higher.

```python
#heat map of correlation for defaulters.
plt.figure(figsize=(12,8)) 
sns.heatmap(df_1.corr(), annot=True, cmap="YlGn")
plt.title('Correlation matrix for defaulters(target variable 1)')
plt.show()
```

```python
#finding top 10 correlations
corr1 = df_1.corr()
corr_df1 = corr1.where(np.triu(np.ones(corr1.shape), k=1).astype(np.bool))
corr_df1 = corr_df1.unstack().reset_index().dropna(subset = [0])
corr_df1.columns = ['Var1', 'Var2', 'Corr_Value']
corr_df1.sort_values(by = "Corr_Value", ascending =False, inplace = True)
corr_df1.head(10)
```

## Inference from target 1

* AMT_GOODS_PRICE and AMT_CREDIT are highly correlated, which means that when credit amount of the loan is higher the amount of goods price for which loan is given  is also higher.
* Second highest correlation is seen between AMT_GOODS_PRICE and AMT_ANNUITY which implies that for higher amount of goods price loan annuity is also higher.
* Third highest correlation is seen between AMT_ANNUITY and	AMT_CREDIT	which implies for higher loan annuity, credit amount of the loan is also higher.

```python

apprvd_df = merged_df[merged_df['NAME_CONTRACT_STATUS']=='Approved']
refsd_df = merged_df[merged_df['NAME_CONTRACT_STATUS']=='Refused']
cncld_df = merged_df[merged_df['NAME_CONTRACT_STATUS']=='Canceled']
unusd_df = merged_df[merged_df['NAME_CONTRACT_STATUS']=='Unused offer']
```

```python
#we are defining a function
def full_plot(variable_name):
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,ncols=2, figsize=(15,12), sharey='all')
    fig.tight_layout(pad=10.0)

   
    ax1 = sns.countplot(ax=ax1,data=refsd_df,x=variable_name)
    ax1.set_title("Refused Loan", fontsize=10)
    ax1.set_ylabel("Number of Loans")
    ax1.set_xticklabels(ax1.get_xticklabels(),rotation=90)
    
    ax2 = sns.countplot(ax=ax2,data=apprvd_df,x=variable_name)
    ax2.set_title("Approved Loan", fontsize=10)
    ax2.set_ylabel("Number of Loans")
    ax2.set_xticklabels(ax2.get_xticklabels(),rotation=90)

    ax3 = sns.countplot(ax=ax3,data=cncld_df,x=variable_name)
    ax3.set_title("Canceled Loan", fontsize=10)
    ax3.set_xlabel(variable_name)
    ax3.set_ylabel("Number of Loans")
    ax3.set_xticklabels(ax3.get_xticklabels(),rotation=90)

    ax4 = sns.countplot(ax=ax4,data=unusd_df,x=variable_name)
    ax4.set_title("Unused Loan", fontsize=10)
    ax4.set_xlabel(variable_name)
    ax4.set_ylabel("Number of Loans")
    ax4.set_xticklabels(ax4.get_xticklabels(),rotation=90)
    
    plt.show()
```

```python
full_plot('NAME_CONTRACT_TYPE_y')
```

### Inference
* A large number of cash loans gets refused where as only a small number of revolving loans gets refused.
* Among approved Consumer loans the client doesn't cancel it but thier exist some unused consumer loans.
* Cash loans are largest among refused and second largest among approved cases but in approved cases many clients cancel it and rest of the clients uses it that means no any cash loan stays unused.
* Revolving loans are third largest among refused and approved category but  many clients cancel it and rest of the clients uses it that means no any cash loan stays unused.
